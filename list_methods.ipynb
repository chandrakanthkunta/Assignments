# List


A list in Python is a built-in data structure used to store a collection of items. Lists are ordered, mutable (changeable), and can contain elements of different data types, including numbers, strings, and other objects. Lists are defined by enclosing a comma-separated sequence of values within square brackets [].

```python
my_list = [1, 2, 3, "apple", "banana", True]
my_list
```

##### empty list 

```python
a = []
a
```

### Indexing and Slicing in Lists

```python
list1 = ["apple", "banana", "cherry"] 
list2 = ["abc", 34, True, 40, "male"]
print(list2[0], list2[-1]) #indexing
print(list2[::], list2[::-1], sep ='\n') #slicing

print(list1, type(list1), id(list1), sep ='\n')
print(list2, type(list2), id(list2), sep ='\n')
```


# list()

```python
b = list()
print(b,len(b),hex(id(b)) ,sep="\n")
```

##### individual

```python
c = list("chandu")
print(c[0])
print(c)
```

# Nested list()

```python
d = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(d[0])
print(d)
```

# Range

the range() function is used to generate a sequence of numbers within a specified range. It's often used in loops and for creating sequences of integers.



range(start, stop): Generates a sequence of integers starting from the start value up to (but not including) the stop value

```python
e = range(0,5)
print(range(10))
```

##### append()

 the append() method is used to add an element to the end of a list. It is a commonly used method for modifying lists by adding new elements

```python
help(list.append)
```

```python
my_list = [1, 2, 3, 3, 4, 5, 6, 1]

my_list.append(4)  

print(my_list) 
```

#### count()

the count() method is used to count the number of occurrences of a specified element in a list. This method is helpful when you want to determine how many times a particular element appears in the list

```python
my_list.count(3)
```

##### insert()

the insert() method is used to insert an element at a specified position within a list. This method allows you to add an element at a specific index while shifting the existing elements to make room for the new one. 

```python
my_list.insert(2,9)
my_list
```

```python
my_list.insert(-3,"list")
my_list
```

##### pop()

the pop() method is used to remove and return an element from a list at a specified index. The method modifies the original list by removing the element, and it also returns the removed element. 

```python
help(list.pop)
```

```python
my_list.pop(0)
```

```python
my_list.pop(-1)
```
