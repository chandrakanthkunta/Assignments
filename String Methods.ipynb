# count()

##### The count() method is a built-in method in Python used to count the number of occurrences of a specified substring within a given string. 

```python
help(str.count)
```

```python
text = '''Betty Botter bought some butter
But she said the butter’s bitter
If I put it in my batter, it will make my batter bitter
But a bit of better butter will make my batter better
So ‘twas better Betty Botter bought a bit of better butter'''

count = text.count("butter")
print(count)  # Output: 3

# You can specify a start and end index for the counting
count = text.count("Botter", 5, 50)  # Count within the slice [5:50]
print(count)  # Output: 1
```

# find()

##### The find() method is a string method in Python used to find the index of the first occurrence of a substring within a given string. It returns the index of the first character of the first occurrence of the substring, or -1 if the substring is not found

```python
help(str.find)
```

```python


# Find the index of "put"in the text
index = text.find("put")
print(index)  # Output: 70

# If the substring is not found, find() returns -1
index = text.find("Python")
print(index)  # Output: -1

# You can specify a start and end index for the search
index = text.find("she", 12, 80)  # Start searching from index 12 to 30
print(index)  # Output: 36
```

# rfind()

##### the rfind() method is a string method used to find the last occurrence of a specified substring within a given string. This method is similar to the find() method but searches for the substring from the end of the string to the beginning. It returns the index of the first character of the last occurrence of the substring or -1 if the substring is not found. 

```python
help(str.rfind)
```

```python

# Find the last index of "butter" in the text
index = text.rfind("butter")
print(index)  # Output: 227

# If the substring is not found, rfind() returns -1
index = text.rfind("cheese")
print(index)  # Output: -1

# You can specify a start and end index for the search
index = text.rfind("butter", 10, 50)  # Start searching from index 10 to 50
print(index)  # Output: -1 (not found in the specified range)
```

# split()

###### the split() method is a string method that is used to split a string into a list of substrings based on a specified delimiter. 

```python
help(str.split)
```

```python

# Split the text into a list of words (default delimiter is whitespace)
words = text.split()
print(words)


# Split the text using a custom delimiter (",")
csv_data = "python,version,3.7,New York"
csv_values = csv_data.split(",")
print(csv_values)


# Specify a maximum number of splits (2)
sentence = "Python is fun to learn and Python is versatile."
fragments = sentence.split("Python", 2)
print(fragments)

```

# rsplit()

###### the rsplit() method is similar to the split() method, but it splits a string from the right (end) into a list of substrings based on a specified delimiter.

```python
help(str.rsplit)
```

```python

# Split the text into a list of words from the right (default delimiter is whitespace)
words = text.rsplit()
print(words)

# Split the text using a custom delimiter ("i")
parts = text.rsplit("i")
print(parts)


# Specify a maximum number of splits (2)
sentence = "Python is fun to learn and Python is versatile."
fragments = sentence.rsplit("Python", 2)
print(fragments)

```

# partition()

###### the partition() method is a string method that splits a string into three parts based on the first occurrence of a specified separator. 

```python
help(str.partition)
```

```python

# Split the text into three parts using the first space as a separator
(head, separator, tail) = text.partition(" ")
print("Head:", head)  # Output: "Python"
print("Separator:", separator)  # Output: " "
print("Tail:", tail)  # Output: "is a versatile programming language."

# Split the text into three parts using the first occurrence of "is" as a separator
(head, separator, tail) = text.partition("is")
print("Head:", head)  # Output: "Python "
print("Separator:", separator)  # Output: "is"
print("Tail:", tail)  # Output: " a versatile programming language."
```

# rpartition()

###### the rpartition() method is similar to the partition() method, but it splits a string into three parts based on the last occurrence of a specified separator. 

```python
help(str.rpartition)
```

```python
text = "Python is a versatile programming language. Python is fun to learn."

# Split the text into three parts using the last space as a separator
(head, separator, tail) = text.rpartition(" ")
print("Head:", head)  # Output: "Python is a versatile programming language. Python is fun to"
print("Separator:", separator)  # Output: " "
print("Tail:", tail)  # Output: "learn."

# Split the text into three parts using the last occurrence of "Python" as a separator
(head, separator, tail) = text.rpartition("Python")
print("Head:", head)  # Output: "Python is a versatile programming language. "
print("Separator:", separator)  # Output: "Python"
print("Tail:", tail)  # Output: " is fun to learn."
```

# splitlines()

###### the splitlines() method is a string method used to split a string into a list of substrings based on line breaks or newline characters ('\n') in the string.

```python
help(str.splitlines)
```

```python
text = "This is line 1.\nThis is line 2.\nThis is line 3."

# Split the text into a list of lines without keeping line endings
lines = text.splitlines()
print(lines)
# Output: ['This is line 1.', 'This is line 2.', 'This is line 3.']

# Split the text into a list of lines while keeping line endings
lines_with_endings = text.splitlines(keepends=True)
print(lines_with_endings)
# Output: ['This is line 1.\n', 'This is line 2.\n', 'This is line 3.']

# Split a text with Windows-style line endings ("\r\n")
text_windows = "Line 1.\r\nLine 2.\r\nLine 3."
lines_windows = text_windows.splitlines()
print(lines_windows)
# Output: ['Line 1.', 'Line 2.', 'Line 3.']
```
